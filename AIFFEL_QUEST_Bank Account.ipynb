```python
# [QUEST 03: 은행계좌 만들기]
# 조건:
# - Account 클래스
# ◦ 은행이름, 예금주, 계좌번호, 잔액 설정
# ◦ 은행이름은 '모두은행'
# ◦ 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성
# ◦ 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수 저장

# - 입금 메서드
# ◦ 입금은 최소 1원 이상만 가능
# ◦ 입금 횟수가 5회가 될 때 잔고 기준으로 1% 이자 지급
# ◦ 입금 내역 기록

# - 출금 메서드
# ◦출금은 계좌의 잔고 이상으로 할 수 없음
# ◦출금 내역 기록

# - get account num() 메서드
# ◦ Account 클래스로부터 생성된 계좌 개수 출력

# - history() 메서드
# ◦입금 내역과 출금 내역을 출력
```

```python
import random
import datetime

# 계좌 관리 클래스
class Account:
    # 이름은 '모두은행', 계좌 총 개수 = 0
    BANK_NAME = '모두은행'
    account_count = 0

    def __init__(self, customer):
        Account.accuont_count += 1

        self.customer = customer # 고객 이름
        self.bank_name = Account.BANK_NAME # 은행 이름
        self.account_nember = self.rand_account_number() # 랜덤 생성된 계좌 번호
        self.money = 0 # 현재 잔고
        self.income_count = 0 # 입금 횟수
        self.history_list= [] # 입출금 리스트

    def rand_account_number(self):
        """3자리-2자리-6자리 형식의 랜덤 계좌번호를 생성합니다."""
        part1 = str(random.randint(0,999)).zfill(3)
        part2 = str(random.randint(0,99)).zfill(2)
        part3 = str(random.randint(0,999999)).zfill(6)
        return f'{part1}-{part2}-{part3}'

    # 입금 메서드
    def deposit(self, amount):
        if amount < 1:
            print("입금은 1원 이상부터 가능합니다.")
            return

        self.money += amount
        self.income_count += 1
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        self.history_list.append(f"[{now}] 입금: +{amount}원, 잔액: {self.money}원")
        print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.money}원")

        # 입금 5회마다 잔액기준 1% 이자 지급
        if self.income_count % 5 == 0:
            interest = int(self.money * 0.01)
            self.money += interest
            self.history_list.append(f"[{now}] 입금 횟수 5회 달성! 이자 +{interest}원 지급, 최종 잔액: {self.money}원")
            print(f"축하합니다! 입금 횟수 5회 달성으로 {interest}원의 이자가 지급되었습니다.")

    # 출금 메서드
    def withdraw(self, amount):
        """
        계좌에서 출금하는 메서드입니다.
        잔액보다 많은 금액은 출금할 수 없습니다.
        """
        if amount > self.money:
            print(f"출금 실패: 잔액이 부족합니다. (현재 잔액: {self.money}원)")
            return

        self.money -= amount
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        self.history_list.append(f"[{now}] 출금: -{amount}원, 잔액: {self.money}원")
        print(f"{amount}원 출금 완료. 현재 잔액: {self.money}원")

    @classmethod
    def get_account_num(cls):
        """생성된 계좌 객체의 총 개수를 출력합니다."""
        print(f"총 {cls.accont_count}개의 계좌가 생성되었습니다.")

    def history(self):
        """입금 및 출금 내역을 모두 출력합니다."""
        print(f"\n[{self.customer}]님의 계좌 내역 (계좌번호: {self.account_nember})")
        print("--------------------------------------------------")
        for record in self.history_list:
            print(record)
        print("--------------------------------------------------")
        print(f"최종 잔액: {self.money}원")

if __name__ == "__main__":
    # 사용자로부터 정보를 입력받아 계좌를 생성합니다.
    print("은행 계좌 관리 프로그램에 오신 것을 환영합니다!")
    customer_name = input("고객님 이름을 입력해 주세요: ")
    my_account = Account(customer_name)
    print(f"\n{my_account.customer}님, 계좌가 성공적으로 개설되었습니다.")

    while True:
        print("\n--- 메뉴 ---")
        print("1. 입금")
        print("2. 출금")
        print("3. 내역 조회")
        print("4. 종료")

        try:
            choice = int(input("원하는 메뉴 번호를 입력해 주세요: "))
        except ValueError:
            print("잘못된 입력입니다. 숫자(1, 2, 3, 4)를 입력해 주세요.")
            continue

        if choice == 1:
            try:
                amount = int(input("입금할 금액을 입력해 주세요: "))
                my_account.deposit(amount)
            except ValueError:
                print("잘못된 금액입니다. 숫자를 입력해 주세요.")

        elif choice == 2:
            try:
                amount = int(input("출금할 금액을 입력해 주세요: "))
                my_account.withdraw(amount)
            except ValueError:
                print("잘못된 금액입니다. 숫자를 입력해 주세요.")

        elif choice == 3:
            my_account.history()

        elif choice == 4:
            print("프로그램을 종료합니다.")
            break

        else:
            print("존재하지 않는 메뉴 번호입니다. 다시 입력해 주세요.")
```
